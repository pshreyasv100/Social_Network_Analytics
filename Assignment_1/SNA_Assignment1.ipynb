# Analysis Part 1:
- From the above centrality correlation scores, it is evident that the scores obtained by all the centrality measures are very similar to one another, apart from the similarity between eigen vector centrality and betweenness centrality. This further emphasizes the difference between the two centrality scores because betweenness centrality corresponds to how many shortest paths a node belongs to whereas eigen vector centrality also considers the importance of the nodes that a node is connected to, hence providing a better measure of a node's centrality measure.

# Character Study:

- From the centrality scores for Valjean, who has a high betweenness, degree and closeness centrality score but a relatively lower eigen vector centrality score, it is clear that he is a central actor tied to a lot of other actors, however the other actors he is tied to are not that important and not as important as he is. 

- From the centrality scores for Cosette, who has a very low betweenness, degree and eigen vector centrality score but a relatively higher closeness centrality score, it is evident that although she is not connected to many characters in the book, she is tied to many important actors, such as Valjean and Marius.

- Similar to Cosette, Myriel also has a much higher closeness centrality compared to other other centrality scores, but has a much lower eigen vector (0.02) centrality score than degree and betweenness centrality. This means that Myriel does not have key ties, in spite of a high closeness centrality (because of the very low eigen vector centrality), but maybe because Myriel has multiple ties, but so do many other actors.

-  Marius has a high closeness centrality as well, a lower degree and eigen vector centrality, and an even lower betweenness centrality. This means that although Marius is close to a lot of key actors, his connections are not very important and information bypasses him (for example, he does not know about how Valjean tried to save his life until much later).

```python
def get_percolated_cliques(G, k):
    cliques = list(frozenset(c) for c in nx.find_cliques(G) if len(c) >= k)

    perc_graph = nx.Graph()
    perc_graph.add_nodes_from(cliques)
    for c1, c2 in combinations(cliques, 2):
        if len(c1.intersection(c2)) >= (k - 1):
            perc_graph.add_edge(c1, c2)

    for component in nx.connected_components(perc_graph):
        yield(frozenset.union(*component))

k = 3
mygenerator = get_percolated_cliques(les,k)
for i in mygenerator:
    print(i)
```
