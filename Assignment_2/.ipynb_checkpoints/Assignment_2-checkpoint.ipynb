```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys
import pandas as pd
import networkx as nx
from operator import itemgetter
from itertools import combinations
import re
import pprint
import community
import collections
```

```python
file = open("Adventures_of_Tom_Sawyer.txt", "r") 

text = file.read()
text = text.replace('\n',' ')
text = text.replace("\'",'')
#text = text.replace("-",' ')
#text = text.lower()
print(text[:70])
```

```python
words_list = text.split(' ')
```

```python
'''
characters = ["Tom Sawyer","Aunt Polly","Huckleberry Finn","Becky Thatcher","Joe Harper","Sid","Mary","Injun Joe","Muff Potter",
             "Dr. Robinson","Mr. Sprague","The Widow Douglas","Mr. Jones","Judge Thatcher","Jim","Amy Lawrence","Ben Rogers",
             "Alfred Temple","Mr. Walters","Mr. Dobbins"]
'''
#Using only first names of characters since they are used more commonly than full names in the story.
characters = ["Tom","Aunt Polly","Huck","Becky","Joe","Sid","Mary","Injun Joe","Potter",
             "Robinson","Mr. Sprague","Widow Douglas","Mr. Jones","Judge","Jim","Amy","Ben",
             "Alfred Temple","Mr. Walters","Mr. Dobbins"]
```

### Choice of Window size

##### Pages: 230 pages
##### Wordcount: 63,510 words
##### Window size : Wordcount / Pages ~ 280  * 3 = 840
##### Allowing for a Window Size of 840= number of words in 3 pages so that all actors are connected in the adjacency matrix.

```python
links = []
window =840
window_parts = []
for k in range(0, len(words_list),window):
    text_current = words_list[k:k+window]
    window_parts.append(' '.join(text_current))


for i in range(0,len(characters)):
    for j in range (i+1, len(characters) ):
        count = 0
        for part in window_parts:
            if (len(re.findall(characters[i],part)) > 0) and (len(re.findall(characters[j],part)) > 0):
                count += 1
                    
        edge = []
        edge.append(characters[i])
        edge.append(characters[j])
        edge.append(count)
        links.append(edge)
```

```python
g = nx.Graph()
g.add_nodes_from(characters)
for i in links:
    if (i[2]!= 0):
        g.add_edge(i[0],i[1], weight = int(i[2]))
print(nx.info(g))
```

```python
plt.figure(3,figsize=(16,16)) 
nx.draw_networkx(g,node_color='yellow',with_labels=True)
plt.show()
```

# Community Detection using  Louvain and Clique percolation methods 

```python
def get_percolated_cliques(G, k):
    cliques = list(frozenset(c) for c in nx.find_cliques(G) if len(c) >= k)

    perc_graph = nx.Graph()
    perc_graph.add_nodes_from(cliques)
    for c1, c2 in combinations(cliques, 2):
        if len(c1.intersection(c2)) >= (k - 1):
            perc_graph.add_edge(c1, c2)

    for component in nx.connected_components(perc_graph):
        yield(frozenset.union(*component))

k = 5
mygenerator = get_percolated_cliques(g,k)
for i in mygenerator:
    print(i)
```

```python
#Random graph model
#No. of nodes=20, p=0.62 because density of the graph is 0.62
plt.figure(3,figsize=(12,12)) 
random = nx.gnp_random_graph(20,0.62)
nx.draw_circular(random, with_labels = True,font_size = 8)
plt.draw()
```

```python
#Watts strogatz small world graph.
#k=12 because average degree for a node in original graph is 11.8
#A p=0.8 is chosen because we want some nodes to have more edgs than others.
plt.figure(3,figsize=(12,12)) 
sml = nx.watts_strogatz_graph(20,12,0.8)
nx.draw_circular(sml, with_labels = True,font_size = 8)
plt.draw()
```

```python
#Barabasi-Albert Preferential Attachment Graph.
#No. of nodes = 20, m=12 because average degree is 11.8
plt.figure(3,figsize=(12,12)) 
pref = nx.barabasi_albert_graph(20,12)
nx.draw_circular(pref, with_labels = True,font_size = 8)
plt.draw()
```
